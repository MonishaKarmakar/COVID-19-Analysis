#### **OBJECTIVE-** TO analyse the COVID-19 spread in India i.e. on 21st OCTOBER 2020

#### **The Dataset is taken from the website https://www.mohfw.gov.in/data/datanew.json** 

```python
#IMPORTING ALL REQUIRED LIBARIES AND THE DATASET
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
```

```python
data= pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
data
```

```python
#The last observation is not assinged with any name hence we can remove that particular row and the serial numbers that are assinged
data=data.iloc[:-1,1:]
data.head()
```

#### Performing expolatory Data Analysis
#### Total number of positive cases state wise


```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='black')
plt.barh(total_states,data['positive'], align='center', alpha=0.6,  
                 color=(0.5,0.2,0.9),  
                 edgecolor=(0.5,0.2,0.5))
    
plt.yticks(total_states, data['state_name'])  
plt.xlim(1,max(data['positive'])+100) #so that we have extra 100 spacing in the axis
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()
```

#### Total number of death recorded state wise

```python
#plotting total death reported
figure(num=None,figsize=(10,9),dpi=80,facecolor="w",edgecolor='black')
plt.barh(total_states,data['death'],align="center",alpha=0.5,
        color=(0.2,0.3,0.1),edgecolor=(0.5,0.4,0.1))
plt.yticks(total_states, data['state_name'])
plt.xlim(1,max(data['death']+100))
plt.xlabel("Death Numbers")
plt.title("DEATH RECOREDS")
plt.show();
```

```python
#Total Cured Analysis
figure(num=None,figsize=(10,9),dpi=80,facecolor='w',edgecolor='black')
plt.barh(total_states,data["cured"],align='center', alpha=0.5, color=(0.2,0.8,0.1),edgecolor=(0.1,0.5,0.3))
plt.yticks(total_states,data['state_name'])
plt.xlim(1,max(data['cured']+100))
plt.xlabel("Cases Cured")
plt.title("Cured Cases")
plt.show()

```

##### Total number of Active cases

```python
figure(num=None,figsize=(10,9),dpi=80,facecolor='w',edgecolor="black")
plt.barh(total_states,data["active"],align='center',color=(0.9,0.7,0.1),edgecolor=(0.1,0.5,0.3))
plt.yticks(total_states,data['state_name'])
plt.xlim(1,max(data['active']+100))
plt.title("Active Cases")
plt.xlabel("Active number")
plt.show()
```
